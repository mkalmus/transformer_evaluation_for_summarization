```python
import csv
import pandas as pd
from tqdm.contrib import tzip

import nltk
from nltk.tokenize import sent_tokenize

from datasets import load_dataset
```

```python
nltk.download('punkt')
```

```python
DATA_CACHE_PATH = './cached_data'

pubmed = load_dataset("scientific_papers", "pubmed", split="test", cache_dir=DATA_CACHE_PATH)
# patents = load_dataset('big_patent', 'h', 'test', cache_dir=DATA_CACHE_PATH)
# news = load_dataset('ccdv/cnn_dailymail', '3.0.0', 'test')
```

```python
def three_sentence_summary(text):
    """
    Simple three-sentence extractive baseline
    """
    return "\n".join(sent_tokenize(text)[:3])
```

```python
def get_articles_abstracts(dataset, dataset_name):
    """
    Gets the input and output strings for a dataset based on different column. 
    Assumes you're only working with one split rather than the whole dataset object
    """
    if dataset_name == 'pubmed':
        articles = list(dataset['article'])
        abstracts = list(dataset['abstract'])
    elif dataset_name == 'big_patent':
        articles = list(dataset['description'])
        abstracts = list(dataset['abstract'])
    elif dataset_name == 'cnn_dailymail':
        articles = list(dataset['article'])
        abstracts = list(dataset['highlights'])
    
    return articles, abstracts
```

```python
def create_write_baseline(articles, abstracts):
    with open('baseline_pubmed.csv', 'w') as csvfile:
        
        writer = csv.writer(csvfile)
        writer.writerow(['ground_truth', 'lead3'])

        for art, ab in tzip(articles, abstracts):          
            summary = three_sentence_summary(str(art))
            writer.writerow([ab, summary])
    return None
```

```python
articles, abstracts = get_articles_abstracts(pubmed, 'pubmed')
```

```python
create_write_baseline(articles, abstracts)
```
